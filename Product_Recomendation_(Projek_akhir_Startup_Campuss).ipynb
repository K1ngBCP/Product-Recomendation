```python
def build_chart(genre,percentile=0.75):
    df = gen_df[gen_df['genre'] == genre]
    # Menghitung Quartil
    Quartil = df['total_click'].quantile(percentile)

    # Memilih data yang memiliki total_click di atas atau sama dengan Quartil
    qualified = df[df['total_click'] >= Quartil][['product_id', 'gender', 'subCategory', 'articleType', 'baseColour', 'season', 'year', 'usage', 'productDisplayName', 'total_click', 'quantity', 'item_price', 'total_sale']]

    # Menentukan fungsi weighted_score
    def weighted_score(x):
        w_clicks = 0.4
        w_sales = 0.4
        w_revenue = 0.2
        a = x['total_click']
        b = x['quantity']
        c = x['total_sale']
        return (w_clicks * a + w_sales * b + w_revenue * c)

    # Mengaplikasikan fungsi weighted_score pada DataFrame qualified
    qualified['ws'] = qualified.apply(weighted_score, axis=1)

    # Mengurutkan DataFrame berdasarkan nilai ws secara menurun dan mengambil 250 baris teratas
    qualified = qualified.sort_values('ws', ascending=False).head(250)

    return qualified
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')

# Tokenization and preprocessing
stop_words = set(stopwords.words('english'))

def tokenize_and_preprocess(text):
    tokens = word_tokenize(text.lower())
    tokens = [token for token in tokens if token.isalpha() and token not in stop_words]
    return ' '.join(tokens)

df['processed_description'] = df['productDisplayName'].apply(tokenize_and_preprocess)
```

```python
# Bobot untuk setiap variabel
w_clicks = 0.4
w_sales = 0.4
w_revenue = 0.2

# Hitung Weighted Score
df['weighted_score'] = (w_clicks * df['total_click'] + w_sales * df['quantity'] + w_revenue * df['total_sale'])
```

```python
df
```

```python
# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

# Reshape kolom 'weighted_score' menjadi bentuk array satu dimensi (diperlukan oleh MinMaxScaler)
sales_values = df['weighted_score'].values.reshape(-1, 1)

# Fit dan transformasi data menggunakan MinMaxScaler
scaled_sales = scaler.fit_transform(sales_values)

# Mengganti nilai dalam kolom 'weighted_score' dengan nilai yang telah di-skala
df['weighted_score'] = scaled_sales
```

```python
# TF-IDF vectorization
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['processed_description'])
```

```python
# Calculate cosine similarity
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# Function to get recommendations with product IDs
def get_recommendations_with_ids(product_id, k=5):
    item_index = df[df['product_id'] == product_id].index[0]
    sim_scores = list(enumerate(cosine_similarities[item_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:k+1]  # Exclude the item itself (index 0)
    similar_items = [i[0] for i in sim_scores]

    # Return both product names and product IDs
    recommended_products = df.iloc[similar_items][['productDisplayName', 'product_id']]

    return recommended_products
```

```python
get_recommendations_with_ids(51056)
```

```python
# Function to get recommendations with only 'productId' and 'productDisplayName' columns, sorted by 'ws' and cosine similarity
def get_recommendations_only_columns_sorted(product_id, k=5, ws_weight=0.2, cosine_sim_weight=0.8):
    item_index = df[df['product_id'] == product_id].index[0]
    sim_scores = list(enumerate(cosine_similarities[item_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:k+1]  # Exclude the item itself (index 0)
    similar_items = [i[0] for i in sim_scores]

    # Calculate a combined score for sorting
    combined_scores = [(i, ws_weight * df['weighted_score'].iloc[i] + cosine_sim_weight * sim_scores[j][1])
                       for j, i in enumerate(similar_items)]

    # Sort the similar items based on the combined score
    sorted_similar_items = sorted(combined_scores, key=lambda x: x[1], reverse=True)
    sorted_similar_items = [i[0] for i in sorted_similar_items]

    # Select only the relevant columns
    result_df = df[['product_id', 'productDisplayName']].iloc[sorted_similar_items].reset_index(drop=True)

    return result_df
```

```python
# Function for search
def search_items_df(query):
    query = tokenize_and_preprocess(query)
    query_vector = tfidf_vectorizer.transform([query])
    cosine_similarities_search = linear_kernel(query_vector, tfidf_matrix).flatten()
    result_indices = cosine_similarities_search.argsort()[::-1]

    # Create a DataFrame with the search results
    search_results_df = pd.DataFrame({
        'product_id': df['product_id'].iloc[result_indices],
        'ProductDisplayName': df['productDisplayName'].iloc[result_indices],
        'CosineSimilarity': cosine_similarities_search[result_indices]
    }).head(5)  # Limit to 15 rows

    return search_results_df
```

```python
# Function for search
def search_items_cc(query):
    query = tokenize_and_preprocess(query)
    query_vector = tfidf_vectorizer.transform([query])
    cosine_similarities_search = linear_kernel(query_vector, tfidf_matrix).flatten()
    weighted_score = df['weighted_score']  # Ganti 'weighted_score' dengan nama kolom yang sesuai

    # Calculate the combined score with a weighted approach
    combined_score = 0.8 * cosine_similarities_search + 0.2 * weighted_score

    result_indices = combined_score.argsort()[::-1]

    # Create a DataFrame with the search results and combined scores
    search_results_df = pd.DataFrame({
        'product_id': df['product_id'].iloc[result_indices],
        'ProductDisplayName': df['productDisplayName'].iloc[result_indices],
        'CosineSimilarity': cosine_similarities_search[result_indices],
        'WeightedScore': weighted_score.iloc[result_indices],
        'CombinedScore': combined_score[result_indices]
    }).head(5)  # Limit to 15 rows

    return search_results_df
```

```python
from surprise.model_selection import GridSearchCV
# Tentukan daftar parameter yang akan diuji
param_grid = {'n_factors': [10, 20, 30],
              'n_epochs': [5, 10, 15],
              'lr_all': [0.002, 0.005, 0.01],
              'reg_all': [0.02, 0.1, 0.4]}
```

```python
# Inisialisasi objek GridSearchCV dengan model, parameter grid, dan metrik evaluasi yang diinginkan
grid_search = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=5)
```

```python
# Fitting model ke data dan melakukan grid search
grid_search.fit(dataku)
```

```python
# Menampilkan parameter terbaik
print("Parameter terbaik:", grid_search.best_params)

# Menampilkan nilai RMSE dan MAE terbaik
print("Nilai RMSE terbaik:", grid_search.best_score['rmse'])
print("Nilai MAE terbaik:", grid_search.best_score['mae'])
```

```python
# Retrieve the best parameters
best_params = grid_search.best_params['rmse']

# Inisialisasi objek SVD dengan parameter terbaik
algo = SVD(n_factors=best_params['n_factors'],
           n_epochs=best_params['n_epochs'],
           lr_all=best_params['lr_all'],
           reg_all=best_params['reg_all'])
```

```python
# Hybrid recommendation combining content-based and collaborative filtering
def hybrid_recommendations(user_id, query, k=5):
    # Search for product IDs based on the query
    search_results_df = search_items_df(query)
    if not search_results_df.empty:
      selected_product_id = search_results_df.iloc[0]['product_id']
    else:
      print("No results found for the given query.")

    search_results_df = selected_product_id
    product_ids = search_results_df

    # Content-Based Recommendations
    content_based_rec = get_recommendations_with_ids(product_ids, k)

    # Collaborative Filtering Recommendations
    collaborative_filtering_rec = pd.DataFrame({
        'product_id': content_based_rec['product_id'],
        'est': [algo.predict(user_id, pid, 1.2).est for pid in content_based_rec['product_id']]
    })

    # Combine recommendations based on 'product_id'
    hybrid_rec = content_based_rec.merge(collaborative_filtering_rec, on='product_id', how='outer').drop_duplicates(subset='product_id').reset_index(drop=True)

    return hybrid_rec
```

```python
# Hybrid recommendation combining content-based and collaborative filtering
def hybrid_recommendations(user_id, query, k=4):
    # Search for product IDs based on the query
    search_results_df = search_items_cc(query)
    if not search_results_df.empty:
      selected_product_id = search_results_df.iloc[0]['product_id']
    else:
      print("No results found for the given query.")

    search_results_df = selected_product_id
    product_ids = search_results_df

    # Content-Based Recommendations
    content_based_rec = get_recommendations_only_columns_sorted(product_ids, k)

    # Collaborative Filtering Recommendations
    collaborative_filtering_rec = pd.DataFrame({
        'product_id': content_based_rec['product_id'],
        'est': [algo.predict(user_id, pid, 1.2).est for pid in content_based_rec['product_id']]
    })

    # Combine recommendations based on 'product_id'
    hybrid_rec = content_based_rec.merge(collaborative_filtering_rec, on='product_id', how='outer').drop_duplicates(subset='product_id').reset_index(drop=True)

    # Sort the resulting DataFrame by the 'est' column
    hybrid_rec_sorted = hybrid_rec.sort_values(by='est', ascending=False).reset_index(drop=True)

    return hybrid_rec_sorted
```
